```python
import json
import matplotlib.pyplot as plt

# Danh sch file v nhn tng ng
files = {
    "FedAvg": "results/results_fedavg_dirichlet.json",
    "FedProx": "results/results_fedprox_dirichlet.json",
    "FedAdam": "results/results_fedadam_dirichlet.json"
}

# Mu cho tng thut ton
colors = {
    "FedAvg": "blue",
    "FedProx": "orange",
    "FedAdam": "green"
}

# Khi to dict cha d liu
loss_data = {}
acc_data = {}
epochs = []

# c tng file v trch xut loss + accuracy
for name, filename in files.items():
    with open(filename, "r") as f:
        data = json.load(f)
    
    sorted_epochs = sorted([int(k) for k in data.keys()])
    if not epochs:
        epochs = sorted_epochs  # ch cn ly 1 ln
    loss_data[name] = [data[str(e)]["loss"] for e in sorted_epochs]
    acc_data[name] = [data[str(e)]["cen_accuracy"] for e in sorted_epochs]

# V  th
plt.figure(figsize=(12, 5))

# Loss
plt.subplot(1, 2, 1)
for name in files:
    plt.plot(epochs, loss_data[name], label=name, color=colors[name])
plt.title("Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Accuracy
plt.subplot(1, 2, 2)
for name in files:
    plt.plot(epochs, acc_data[name], label=name, color=colors[name])
plt.title("Accuracy over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
